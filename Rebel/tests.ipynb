## Small version of CFRM

```python
'''
0 is play 1 is sell
'''

p1_h =  np.array([.5,.5])
p1_t = np.array([.5,.5])
p2 = np.array([.5, .5])

def p_heads(ph, pt):
    return(np.array([ph[0]/(ph[0]+pt[0]), 1-ph[0]/(ph[0]+pt[0])]))
    
p_heads(p1_h, p1_t)
```

```python
num_steps = 10000
for i in range(100):
    p1_regret_h = np.array([sum(np.array([-1,1])*p2), .5])
    
    p1_regret_t = np.array([sum(np.array([1,-1])*p2), -.5])

    ph = p_heads(p1_h,p1_t)
    p2_regret = ph
    p1_h = ((1+i)*p1_h + norm(p1_regret_h))/(i+2)
    p1_t = ((1+i)*p1_t + norm(p1_regret_t))/(i+2)
    p2 = ((1+i)*p2 + norm(p2_regret))/(i+2)
    print(f"Iteration {i}")
    print("p1")
    print(norm(p1_regret_h))
    print(norm(p1_regret_t))
    print(p1_h, p1_t)
    print("p2")
    print(p2_regret)
    print(norm(p2_regret))
    print(p2)
    

    
    
```

```python
def norm(policy):
    if sum(policy-min(policy)) ==0:
        return(np.array([.5,.5]))
    ans = (policy-min(policy))/sum(policy-min(policy))
    return(ans)

norm(p1_h)
```
